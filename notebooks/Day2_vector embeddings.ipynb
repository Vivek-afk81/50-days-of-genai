```python
import gensim.downloader as api
model=api.load("word2vec-google-news-300") # Download the model and return as object ready to use
```

Vectors can capture semantic meaning because words that appear in similar contexts end up with similar vectors.

* dog     → [ 0.21,  1.02, -0.33,  0.87,  0.44 ]
* cat     → [ 0.19,  0.98, -0.29,  0.90,  0.41 ]

* apple   → [-1.12, -0.45,  0.88, -0.66,  1.09 ]
* banana  → [-1.05, -0.48,  0.91, -0.62,  1.03 ]

these are 5 dimensional vector space not enough to capture the essence of semantic meaning

we will use :-
word2vec-google-news-300 is a widely used pre-trained word embedding model that represents words as 300-dimensional vectors.

```python
word_vectors=model

print(word_vectors['chicken'])
print(word_vectors['banana'].shape)
```

##Similar Words


KING+WOMAN-MAN=?

```python
results = word_vectors.most_similar(
    positive=['king', 'woman'],
    negative=['man'],
    topn=10
)

for word, score in results:
    print(f"{word}: {score:.4f}")
```

#Checking the similarity between similar pair of words

```python
# Example of calculating similarity

#gender pairs (high similarity)
print(word_vectors.similarity('actor', 'actress'))
print(word_vectors.similarity('father', 'mother'))
print(word_vectors.similarity('husband', 'wife'))
print(word_vectors.similarity('brother', 'sister'))
print("\n")
#Synonyms (very high similarity)
print(word_vectors.similarity('big', 'large'))
print(word_vectors.similarity('small', 'tiny'))
print(word_vectors.similarity('fast', 'quick'))
print(word_vectors.similarity('happy', 'joyful'))
print("\n")
# Weak or unrelated pairs
print(word_vectors.similarity('paper', 'water'))
print(word_vectors.similarity('banana', 'justice'))
print(word_vectors.similarity('car', 'happiness'))
```

```python
print(word_vectors.most_similar('paper',topn=5))
```

###Vector similarity


```python
import numpy as np

# Words to compare
word1 = 'man'
word2 = 'woman'

word3 = 'semiconductor'
word4 = 'earthworm'

word5 = 'hi'
word6 = 'hello'

# Calculate the vector differences
vector_difference1 = model[word1] - model[word2]
vector_difference2 = model[word3] - model[word4]
vector_difference3 = model[word5] - model[word6]

# Calculate the magnitudes of the vector differences
magnitude_of_difference1 = np.linalg.norm(vector_difference1)
magnitude_of_difference2 = np.linalg.norm(vector_difference2)
magnitude_of_difference3 = np.linalg.norm(vector_difference3)

# Print results
print("The magnitude of the difference between '{}' and '{}' is {:.2f}".format(word1, word2, magnitude_of_difference1))
print("The magnitude of the difference between '{}' and '{}' is {:.2f}".format(word3, word4, magnitude_of_difference2))
print("The magnitude of the difference between '{}' and '{}' is {:.2f}".format(word5, word6, magnitude_of_difference3))
```
