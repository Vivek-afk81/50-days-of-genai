```python
from groq import Groq
import os
```

```python
#Setup groq

api_key = os.getenv("GROQ_API_KEY")
client=Groq(api_key=api_key)

def ask(prompt,model="llama-3.1-8b-instant"):
    response=client.chat.completions.create(
        model=model,
        messages=[{"role":"user","content":prompt}],
        temperature=0 # importatnt so that there is fair comparision
    )
    return response.choices[0].message.content
```

## Zero-Shot prompting

Zero‑shot → “No practice runs.”

The AI has never seen the training examples, but it still understands because it has learned patterns from huge amounts of data.


```python
zeroShot ="""A juggler can juggle 16 balls. Half of the balls
are golf balls, and half of the golf balls are blue. How many blue golf balls are there?
"""
print(ask(zeroShot))
```

```python
zeroShot1="""Which word does NOT belong with the
others?
a. couch
b. rug
c. table
d. chair"""
print(ask(zeroShot1))
```

## Few-Shot prompting

Few‑shot → a handful of examples to “teach” the AI 

Few‑shot prompting shows the AI a few examples so it understands the style or task before solving new ones.

```python
fewShotPrompt = """Q: There are 15 trees in a grove. Grove workers
will plant trees in the grove today. After they are done, there will
be 21 trees. How many trees did the grove workers plant?
A: There are 15 trees originally. After planting, there are 21 trees.
So they planted 21 - 15 = 6 trees. The answer is 6.

Q: Roger has 5 tennis balls. He buys 2 more cans of tennis balls.
Each can has 3 tennis balls. How many tennis balls does he have now?
A: Roger started with 5 balls. 2 cans of 3 balls each is 2 x 3 = 6
balls. 5 + 6 = 11. The answer is 11.

Q: A juggler can juggle 16 balls. Half of the balls are golf balls,
and half of the golf balls are blue. How many blue golf balls are
there?
A:"""

print(ask(fewShotPrompt))
```

```python
fewShot1="""Which word does NOT belong with the
others?
a. tape
b. twine
c. cord
d. yarn
Ans: The yarn, twine, and cord are all used for tying.
The tape is not used in the same way

Which word does NOT belong with the
others?
a. guitar
b. flute
c. violin
d. cello
Ans: The guitar, violin, and cello are stringed instruments; the flute is a wind instrument.

Which word does NOT belong with the
others?
a. couch
b. rug
c. table
d. chair"""

print(ask(fewShot1))
```

## Chain of Thought (COT) prompting

Chain of thought = “show your work” for AI.

Chain of Thought is a prompting technique where you ask an AI to show its reasoning step by step instead of just giving the final answer

```python
cotPrompt = """A juggler can juggle 16 balls. Half of the balls are
golf balls, and half of the golf balls are blue. How many blue golf
balls are there?

Let's think step by step."""

print(ask(cotPrompt))
```

```python
cot1="""Which word does NOT belong with the
others?
a. couch
b. rug
c. table
d. chair

show thw reasoning before giving the answer """

print(ask(cot1))
```

```python
#building a comparision table 1

results = {
    "Zero-shot": ask(zeroShot),
    "Few-shot": ask(fewShotPrompt),
    "Chain-of-Thought": ask(cotPrompt)
}

for strategy, output in results.items():
    print(f"\n{'-'*40}\n{strategy}\n{'-'*40}\n{output}")

print()
print("The model is prompt-insensitive on this task ie The task is so easy that all prompting methods reach near-maximum performance.")
```

```python
#building comparision table 2

results1 = {
    "zero Shot": ask(zeroShot1),
    "few Shot": ask(fewShot1),
    "chain of thought": ask(cot1)
}

for strategy, output in results1.items():
    print(f"\n{'-'*40}\n{strategy}\n{'-'*40}\n{output}")
```

## Key Insight 

Prompting strategy matters most exactly when the question
is genuinely ambiguous, not when it's easy. Zero-shot's confident wrong
answer was the most concerning result — it gave no signal that it was
guessing.
